```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import pandas as pd

from plotly.subplots import make_subplots

from sklearn.model_selection import train_test_split, ShuffleSplit, GridSearchCV
from sklearn import preprocessing

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.datasets import fetch_openml
```

```python

data = fetch_openml(name='house_16H', as_frame=True)
```

```python
X = data.data
Y = data.target

scaler = preprocessing.StandardScaler().fit(X)
X_scaled = scaler.transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y,
                                                    random_state=0, train_size=2/3)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)

mse_lin = mean_absolute_error(y_test, y_pred)
print(f'Mean Square Error for line regresor: {mse_lin}')
```

```python
from sklearn.kernel_approximation import Nystroem
from sklearn.linear_model import LinearRegression

from sklearn import pipeline
from sklearn.metrics import mean_squared_error

feature_map_nystroem = Nystroem(random_state=0)
nystroem_approx_krr = pipeline.Pipeline([("feature_map", feature_map_nystroem),
                                        ("LinearRegression", Ridge())])

param_grid = [{'feature_map__gamma': [0.001,0.01,0.1,0.2,0.5,1,2,3,5], 
               'feature_map__n_components': [10,20,30,50,100,500,1000,2000,3000],
               'LinearRegression__alpha' :[0.5,1]}]

grid = GridSearchCV(nystroem_approx_krr, param_grid=param_grid, cv = 3)
grid.fit(X_train,y_train)

y_pred_nystroem = grid.predict(X_test)


mse_nystroem = mean_absolute_error(y_test, y_pred_nystroem)

print('MSE for Nyström: %.4f'%mse_nystroem)
print('MSE (dB) for Nyström: %.4f'%(10*np.log10(mse_nystroem)))
print(f'Parametros: {grid.best_params_}')
```

```python
x_pru = np.arange(len(y_test))
```

```python
fig = make_subplots(rows=2, cols=1,
                    row_heights=[0.2, 0.8],
                    vertical_spacing = 0.02,
                    shared_yaxes=False,
                    shared_xaxes=True)
fig.add_trace(go.Bar(x=x_pru, y = np.abs(y_test-y_pred),name='Diference Linear',marker_color='rgb(208,102,25)', marker_line_color='rgb(108,8,7)',
                  marker_line_width=1.5, opacity=0.6), row=1, col=1)
fig.add_trace(go.Bar(x=x_pru, y = np.abs(y_test-y_pred_nystroem),name='Diference Nystroem',marker_color='rgb(158,202,225)', marker_line_color='rgb(8,8,107)',
                  marker_line_width=1.5, opacity=0.6), row=1, col=1)
fig.update_layout(barmode='relative')

fig.update_yaxes(row=1,col=1,range=[-100,200_000])
fig.add_trace(go.Scatter(x=x_pru,y=y_test, mode='markers',marker_color='rgb(58,102,225)',marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6, name='Test'), row = 2, col=1)
fig.add_trace(go.Scatter(x=x_pru,y=y_pred, mode='markers',marker_color='rgb(208,102,25)', marker_line_color='rgb(108,8,7)',
                  marker_line_width=1.5, opacity=0.6, name='Predicted Linear'),row = 2, col=1)
fig.add_trace(go.Scatter(x=x_pru,y=y_pred_nystroem, mode='markers',marker_color='rgb(58,202,25)',marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6, name='Predicted Nystroem'),row = 2, col=1)
fig.update_layout(width=900, height=600)
fig.update_xaxes(range=[3000,3100])
fig.update_yaxes(row=2,col=1,range=[-100,200_000])

fig.show()
```
